### 1. (b)  Preprocess

> Firstly, we are going to fill the null values with 0. As we observed earlier, the database only contains 2 null records in the variable "children", so it's not a problem to perform these fills.
>
> For "country", "agent", and "company", which also have missing values in the original data as they are categorical, this does not pose a problem as they become base categories during one-hot encoding.

```python
X_train_ohe = X_train_ohe.fillna(0)
X_test_ohe = X_test_ohe.fillna(0)
```

> Then, we proceed to normalize our data before introducing it to ML models to standardize feature scales, facilitate optimization algorithm convergence, and mitigate the impact of outliers, thereby enhancing model performance and interpretation. 

```python
scaler = MinMaxScaler()
X_train_ohe[X_train_ohe.columns] = scaler.fit_transform(
    X_train_ohe[X_train_ohe.columns]
)
X_test_ohe[X_test_ohe.columns] = scaler.transform(
    X_test_ohe[X_test_ohe.columns])
```

> Finally, we proceed to split our data into 80% for training and 20% for validation.

```python
X_train_80, X_val_20, y_train_80, y_val_20 = train_test_split(
    X_train_ohe, y_train, test_size=0.2, random_state=42
)
```

> In our previous EDA, we saw that the original samples with 29 variables were transformed into 940 using one-hot encoding. This is because some of the categorical variables have numerous different categories. Especially notable is the case of the categories country, agent, and company, which have a multitude of different categories, some of which only have a few dozen observations.
> 
> Therefore, after performing some iterations and considering the reduction of processing times for our next models, the decision was made to carry out a feature selection that might seem somewhat arbitrary. This involves eliminating variables with fewer than 50 records. However, this allows us to improve our computational times for variables that likely have little importance in prediction. Additionally, it helps us prevent overfitting, as by reducing the number of variables, we are simplifying the model and avoiding it from fitting too closely to the training data, which could lead to poor generalization on new data. Furthermore, by removing variables with a limited number of records, we are reducing noise in our data and improving the quality of our predictions by focusing on the most relevant and representative variables of the phenomenon we are studying.
>
>In this way, we are left with 224 features out of the original 940. It is worth mentioning that some models such as Random Forest or XGBoost have their own methods of feature selection, but we opted for this alternative initially to select the same number of features for all models and facilitate comparison and processing. In future iterations, we can evaluate more robust feature selections.


```python
selected_features = []
for feature in X_train_80.columns:

    if (X_train_80[feature] != 0).sum() > 50:
        selected_features.append(feature)


X_train_80_selected = X_train_80[selected_features]
X_val_20_selected = X_val_20[selected_features]
X_train_ohe_selected = X_train_ohe[selected_features]
X_test_ohe_selected = X_test_ohe[selected_features]
```

>Random Search was used for hyperparameter selection due to its efficiency. When tuning 5 models with different parameters in a large dataset, times could escalate significantly. Random Search has the advantage of efficiency, allowing for exploration within a broader range than could be achieved with Grid Search in the same time frame, as Grid Search requires testing all combinations.

>For selecting which hyperparameter alternatives to use in the search, the defaults were initially sought, along with a range around them using fractions of the smaller dataset. Some manual experimentation helped understand the range where the most relevant parameters moved.
>
>In the final process, which is presented later, a Random Search with 25 iterations and CV=5 was performed. While more iterations might lead to better results, these 25 iterations yielded superior results compared to the established benchmark for 3 of the models, hence considered an appropriate number given the computational and time constraints.
>
>Below, you can see the training process of the selected models.


```python
# Provisory code to work with a fraction of the data due to efficiency reasons.
n_iter = 25

fraction = 1
n_samples = int(fraction * len(X_train_80_selected))

X_train_80_selected = X_train_80_selected[:n_samples]
y_train_80 = y_train_80[:n_samples]
```

### Logistic Regression

```python
param_grid = {
    "penalty": ["l1", "l2"],
    "C": np.logspace(-4, 4, 20),
    "solver": ["liblinear", "saga"],
}

LR = LogisticRegression(max_iter=1000)

random_search = RandomizedSearchCV(
    LR,
    param_distributions=param_grid,
    n_iter=n_iter,
    scoring="roc_auc",
    cv=5,
    random_state=42,
    n_jobs=-1,
)

random_search.fit(X_train_80_selected, y_train_80)

best_LR = random_search.best_estimator_
best_params_LR = random_search.best_params_

y_pred_proba = best_LR.predict_proba(X_val_20_selected)[:, 1]
auc_val = roc_auc_score(y_val_20, y_pred_proba)

print("Best parameters found:")
print(best_params_LR)
print("Best AUC on validation set:", auc_val)
```

### K Nearest Neighbors

```python
param_grid = {
    "n_neighbors": list(range(1, 21)),
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree", "brute"],
    "leaf_size": list(range(10, 51, 10)),
    "p": [1, 2],
}

KNN = KNeighborsClassifier()

random_search = RandomizedSearchCV(
    KNN,
    param_distributions=param_grid,
    n_iter=n_iter,
    scoring="roc_auc",
    cv=5,
    random_state=42,
    n_jobs=-1,
)

random_search.fit(X_train_80_selected, y_train_80)

best_KNN = random_search.best_estimator_
best_params_KNN = random_search.best_params_

y_pred_proba = best_KNN.predict_proba(X_val_20_selected)[:, 1]

auc_val = roc_auc_score(y_val_20, y_pred_proba)

print("Best parameters found:")
print(best_params_KNN)
print("Best AUC on validation set:", auc_val)
```

### Random Forest

```python
param_grid = {
    "n_estimators": list(range(50, 3001, 200)),
    "max_features": [None, "sqrt", "log2", "auto"],
    "max_depth": list(range(10, 100, 10)),
    "min_samples_split": np.arange(2, 11),
    "min_samples_leaf": np.arange(1, 11),
    "bootstrap": [True, False],
}

RF = RandomForestClassifier()

random_search = RandomizedSearchCV(
    RF,
    param_distributions=param_grid,
    n_iter=n_iter,
    scoring="roc_auc",
    cv=5,
    random_state=42,
    n_jobs=-1,
)

random_search.fit(X_train_80_selected, y_train_80)

best_RF = random_search.best_estimator_
best_params_RF = random_search.best_params_

y_pred_proba = best_RF.predict_proba(X_val_20_selected)[:, 1]
auc_val = roc_auc_score(y_val_20, y_pred_proba)

print("Best parameters found:")
print(best_params_RF)
print("Best AUC on validation set:", auc_val)
```

```python
################################
# Code to plot clusters
################################
def plot_cluster(ax, data, cluster_assignments):
    """Plot two-dimensional data clusters

    Parameters
    ----------
    ax : matplotlib axis
        Axis to plot on
    data : list or numpy array of size [N x 2]
        Clustered data
    cluster_assignments : list or numpy array [N]
        Cluster assignments for each point in data

    """
    clusters = np.unique(cluster_assignments)
    n_clusters = len(clusters)
    for ca in clusters:
        kwargs = {}
        if ca == -1:
            # if samples are not assigned to a cluster (have a cluster assignment of -1, color them gray)
            kwargs = {"color": "gray"}
            n_clusters = n_clusters - 1
        ax.scatter(
            data[cluster_assignments == ca, 0],
            data[cluster_assignments == ca, 1],
            s=5,
            alpha=0.5,
            **kwargs,
        )
        ax.set_xlabel("feature 1")
        ax.set_ylabel("feature 2")
        ax.set_title(f"No. Clusters = {n_clusters}")
        ax.axis("equal")
```

> In the previous plots, we measured dissimilarity using inertia, which is a measure of how much the dispersion of points within each cluster varies. It is calculated as the sum of the squared distances from each point to the nearest centroid in its corresponding cluster. Inertia measures how compact and well-separated the clusters are. The smaller the inertia, the more compact and well-defined the clusters will be.

> When using only two components, the percentage of explained variance is not very high (approximately around 20%). This means that these two components do not capture a large portion of the variability present in the original data. When applying a clustering algorithm in the reduced two-dimensional space, it's possible that the resulting clusters may not be very distinct or well-defined. This is because a significant amount of information is lost during dimensionality reduction, which can make it difficult to clearly separate the data into coherent groups.

### 3 (c) t -SNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
digits_tsne = tsne.fit_transform(X_digits)

plt.figure(figsize=(8, 6))
plt.scatter(digits_tsne[:, 0], digits_tsne[:, 1], c=y_digits, cmap="tab10")
plt.title("Handwritten Digits Visualization with t-SNE")
plt.colorbar(label="Digits")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```

### 3 (d) Strengths and weaknesses of each approach.

> - Comparing both alternatives and their graphs, the t-SNE method seems to have clustered the data more effectively than PCA in the graph. This could be because t-SNE is a nonlinear technique that tends to better preserve local structures and nonlinear relationships between the data, which can lead to clearer clustering in the reduced space.
>
> - The equivalence between the fit and fit_transform methods in t-SNE arises from the integrated nature of the optimization and transformation process in t-SNE, eliminating the need to explicitly separate model fitting and data transformation. However, this integration implies that each model fitting requires a new optimization, which can increase computational complexity and limit the ability to apply the model to new data without retraining it.
